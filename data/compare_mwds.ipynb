```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
peppr_data = pd.read_csv("training_set.csv")
zuo_data = pd.read_csv("zuo_data.csv")
```

```python
peppr_x = peppr_data[["x"+str(i+1) for i in range(15)]]
zuo_x = zuo_data[["x"+str(i+1) for i in range(15)]]
```

```python
xs = [x for x in range(len(peppr_x.iloc[0]))]

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(9,4))
axs[0].boxplot(peppr_x)
axs[0].set_title("PEPPr Train Data")
axs[0].set_ylim(-0.1,1.4)

axs[1].boxplot(zuo_x)
axs[1].set_title("Zuo Data")
axs[1].set_ylim(-0.1,1.4)

for ax in axs:
    ax.yaxis.grid(True)
    ax.set_xticks([y+1 for y in range(15)], labels=["x"+str(i+1) for i in range(15)])
    ax.set_xlabel("Feature")
    ax.set_ylabel("Value")

plt.tight_layout()
plt.savefig("peppr_data_feature_distribution.png")
plt.show()
```

```python
max_x15 = max(list(peppr_x['x15']))
print(max_x15)

within_max_count = 0
for x in list(zuo_x['x15']):
    if x < max_x15:
        within_max_count += 1

print(within_max_count)
print(len(list(zuo_x['x15'])))
```

```python
def mean_pos(data):
    pos = range(len(data))
    return sum((a+1) * b for a, b in zip(pos, data)) / sum(data)

mean_pos_list_zuo = []
for i in range(len(zuo_x)):
    mean_pos_list_zuo.append(mean_pos(zuo_x.iloc[i].to_list()))

mean_pos_list_peppr = []
for i in range(len(peppr_x)):
    mean_pos_list_peppr.append(mean_pos(peppr_x.iloc[i].to_list()))

print(np.mean(mean_pos_list_zuo))
print(np.mean(mean_pos_list_peppr))
```
